```python
"""
Tarea 1 – Extraer el año
------------------------
Objetivo:
- Leer el archivo HTML baby2000.html.
- Extraer el año que aparece en el título "Popularity in XXXX".
- Mostrarlo en pantalla como un string.
"""

import re

# Archivo específico del año 2000
filename = "../data/baby2000.html"

# Leer el contenido
with open(filename, "r", encoding="utf-8") as f:
    texto = f.read()

# Patrón regex: busca "Popularity in " seguido de 4 dígitos
patron = r"Popularity\s+in\s+(\d{4})"
match = re.search(patron, texto)

if match:
    year = match.group(1)
    print("Año:", year)
else:
    print("No se encontró el año en el archivo.")
```

```python
"""
Tarea 2 – Extraer Nombres y Ranking
-----------------------------------
Objetivo:
- Abrir el archivo local baby2000.html.
- Usar regex para extraer ranking, nombre de niño y nombre de niña.
- Guardar cada par (nombre, ranking) en una lista de tuplas.
"""

import re

archivo = "../data/baby2000.html"

# Leer el contenido del archivo
with open(archivo, "r", encoding="utf-8") as f:
    contenido = f.read()

# Patrón regex: número de ranking + nombre niño + nombre niña
patron = r"<td>(\d+)</td><td>([A-Za-z]+)</td><td>([A-Za-z]+)</td>"
coincidencias = re.findall(patron, contenido)

# Armar la lista completa de tuplas
ranking_completo = []
for rank, boy, girl in coincidencias:
    ranking_completo.append((boy, rank))
    ranking_completo.append((girl, rank))

# Mostrar el ranking
print("Total de nombres encontrados:", len(ranking_completo))
print(ranking_completo)
```

```python
"""
Tarea 3 – Guardar en un Diccionario
-----------------------------------
Objetivo:
- A partir de la lista completa de tuplas (nombre, ranking) de la Tarea 2,
  crear un diccionario {nombre: ranking}.
- Si un nombre aparece más de una vez, se conserva la primera aparición.
"""

import re

archivo = "../data/baby2000.html"

# Leer el archivo
with open(archivo, "r", encoding="utf-8") as f:
    contenido = f.read()

# Regex para capturar (rank, boy, girl)
patron = r"<td>(\d+)</td><td>([A-Za-z]+)</td><td>([A-Za-z]+)</td>"
coincidencias = re.findall(patron, contenido)

# Construir el diccionario
diccionario_nombres = {}
for rank, boy, girl in coincidencias:
    for nombre in (boy, girl):
        if nombre not in diccionario_nombres:   # conserva la primera aparición
            diccionario_nombres[nombre] = rank

# Mostrar resultados
print("Cantidad de nombres únicos:", len(diccionario_nombres))
print(diccionario_nombres)
```

```python
"""
Tarea 4 – Lista de Nombres Ordenada
-----------------------------------
Objetivo:
- A partir del diccionario de la Tarea 3,
  obtener una lista con todos los nombres ordenados alfabéticamente.
"""

# Obtenemos todos los nombres del diccionario y los ordenamos
lista_nombres_ordenados = sorted(diccionario_nombres.keys())

# Mostrar resultados
print("Cantidad de nombres ordenados:", len(lista_nombres_ordenados))
print(lista_nombres_ordenados)
```

```python
"""
Tarea 5 – Formatear la Salida Final
-----------------------------------
Objetivo:
- Combinar el año (string) con los nombres ordenados y el diccionario de rankings.
- Crear una lista final:
    * Primer elemento → el año como string.
    * Luego → "nombre - ranking" para cada nombre, ordenados alfabéticamente.
"""

# Construir la lista final
lista_final = [year] + [f"{nombre} - {diccionario_nombres[nombre]}"
                        for nombre in lista_nombres_ordenados]

# Mostrar resultados
print("Cantidad de elementos en la lista final:", len(lista_final))
print(lista_final)
```
